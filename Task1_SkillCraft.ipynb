```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as st
import seaborn as sns
sns.set(style="darkgrid")
```

```python
data = pd.read_csv(r'C:\Users\shrey\OneDrive\Desktop\notable_companies (1).csv')
```

```python
industry_counts = data['Industry'].value_counts().reset_index()
industry_counts.columns = ['Industry', 'Count']
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Industry', y='Count', data=industry_counts, palette="viridis", edgecolor='black',hue='Industry',legend=False)
plt.xticks(rotation=45)
plt.title("Number of Companies by Industry", fontsize=16)
plt.xlabel("Industry", fontsize=14)
plt.ylabel("Number of Companies", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
status_counts = data['Active/Defunct'].value_counts()
sns.barplot(x=status_counts.index, y=status_counts.values, palette="deep", edgecolor='black',hue=status_counts.index,legend=False)
plt.title("Count of Active vs Defunct Companies", fontsize=16)
plt.xlabel("Company Status", fontsize=14)
plt.ylabel("Number of Companies", fontsize=14)
plt.tight_layout()
plt.show()
```
